```python
# Write a NumPy program to implement following operation 
# to add, subtract, multiply, divide arguments element-wise

# Developed By : Pankhaniya Dixant
# Enrollement No : 216270307060


import numpy as np
a = np.array([10,20,30])
print("Array before adding\n",a)
add = np.add(a,50)
print("\nAfter adding values into array: ")
print(add)
sub = np.subtract(add,50)
print("\nAfter subracting values into array: ")
print(sub)
multi= np.multiply(sub,50)
print("\nAfter multiplying values into array: ")
print(multi)
div=np.divide(multi,50)
print("\nAfter dividing values into array: ")
print(div)
```

```python
# Write a NumPy program to implement following operation 
# to round elements of the array to the nearest integer

# Developed By : Pankhaniya Dixant
# Enrollement No : 216270307060

import numpy as np
a = np.array([5.5,10.6,4.3,9.4])
print("Before Round the values: ")
print(a)
b=np.round(a)
print("\nAfter round the values are: ")
print(b)
```

```python
# Write a NumPy program to implement following operation 
# to calculate mean across dimension, in a 2D numpy array

# Developed By : Pankhaniya Dixant
# Enrollement No : 216270307060

import numpy as np
a= np.array(([10,20],[40,50]))
print("Before Mean")
print(a)
print("Mean of each column:")
print(a.mean(axis=0))
print("Mean of each row:")
print(a.mean(axis=1))
```
