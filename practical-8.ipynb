```python
#Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.

# Developed By: Pankhaniya Dixant 
# Enrollement No: 216270307060

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/adaniportys-cvs/ADANIPORTS.csv')

df['Date'] = pd.to_datetime(df['Date'])

start_date = '2007-11-27'
end_date = '2008-02-06'

filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

plt.figure(figsize=(12, 6))
plt.plot(filtered_df['Date'], filtered_df['Open'], label='Opening Price', color='blue')
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Closing Price', color='red')

plt.title('Stock Prices between {} and {}'.format(start_date, end_date))
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.show()
```
