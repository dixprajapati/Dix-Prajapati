```python
# Write a numpy program to implement following operation 
# to create a 3x3 matrix with values ranging from 2 to 10 

# Developed By: Pankhaniya Dixant 
# Enrollement No: 216270307060

import numpy as np
x =  np.arange(2, 11)
y =  x.reshape(3,3)
print(y)
```

```python
# Write a numpy program to implement following operation 
# to append values at the end of an array 

# Developed By: Pankhaniya Dixant 
# Enrollement No: 216270307060

import numpy as np
x = [90, 80, 70]
print("Original array:")
print(x)
x = np.append(x, [[60, 50, 40], [30, 20, 10]])
print("\nAfter append array:")
print(x)
```

```python
# Write a numpy program to implement following operation 
# to create another shape from an array without changing its data(3*2 to 2*3) 

# Developed By: Pankhaniya Dixant 
# Enrollement No: 216270307060

import numpy as np
original = np.array([[1, 2],[3, 4],[5, 6]])


reshaped = original.reshape(2, 3)

print("3x2 Array:")
print(original)

print("\n2x3 Array:")
print(reshaped)
```
